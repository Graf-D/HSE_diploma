```python
import ast

def extract_title(x):
    try:
        return ast.literal_eval(x)[0].lower()
    except:
        # там есть не только строки-списки, но и тупо слово title
        # других не-строк-списоков нет
        return "UNKNOWN"
```

```python
df["title"] = df["title"].apply(extract_title)
df.head()
```

```python
def subj_to_list(x):
    # в subject конечно же тоже лежат строки-списки
    try:
        return ast.literal_eval(x)
    except:
        # и конечно же тоже падает на просто евале
        print(x)
        return ["INCORRECT"]


df["subject"] = df["subject"].apply(subj_to_list)
```

```python
import re
import string


def remove_xml_tags(text):
    pattern = r"<.*?>"
    return re.sub(pattern, " ", text)

def remove_punct(text):
    table = str.maketrans({key: " " for key in string.punctuation})
    return text.translate(table)

def clean_text(text):
    # removes xml tags and punctuation
    # (in that order, bc <> are used to find xml tags)
    # and keeps only one space in a row
    return re.sub(
        "\s+",
        " ",
        remove_punct(
            remove_xml_tags(
                text.lower()
            )
        )
    ).strip()
```

```python
df["cleaned_abstract"] = df["abstract"].apply(clean_text)
```

## Относительная совместная встречаемость

Переделаем матрицу из абсолютной встречаемости в относительную и возможно будем потом как-нибудь использовать для предсказаний

```python
common_occ
```

```python
def relative_freq(row):
    sum_ = np.sum(row)
    new_row = np.zeros(shape=row.shape)
    for i in range(row.shape[0]):
        new_row[i] = row[i] / sum_
    return new_row
```

```python
rel_common_occ = relative_freq(common_occ)
```

```python
def all_fields(subjects):
    fields = []
    for subject in subjects:
        #print(subject)
        try:
            fields.append(subj2field[subj2field["Subject"] == subject]["Field"].values[0])
        except IndexError:
            # Pediatrics, Perinatology, and Child Health
            # почему-то при выкачивании в subj2field разломилось на Pediatrics ; Perinatology and Child Health
            fields.append(subj2field[subj2field["Subject"] == subject.split(",")[0]]["Field"].values[0])
    return fields
```

```python
def field_rate(all_fields, target_field):
    c = Counter(all_fields)
    #return c[target_field] / len(all_fields)
    return 1 if target_field in all_fields else 0
```

# Добавляем колонки subject'ов

```python
df_filtered = pd.read_json("filtered_and_field_1.json")
df_filtered.head()
```

```python
def add_subj(all_subjects, target_subject):
    return 1 if target_subject in all_subjects else 0
```

```python
from itertools import chain
from collections import Counter

subjects = Counter(list(chain(*df_filtered["subject"].values))).most_common(200)
```

```python
len(subjects)
```

```python
from tqdm import tqdm

for subj, _ in tqdm(subjects):
    column_name = "subj_" + subj
    df_filtered[column_name] = df_filtered["subject"].apply(add_subj, args=(subj,))
```
