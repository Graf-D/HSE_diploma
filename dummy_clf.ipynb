```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
```

```python
df = pd.read_pickle("looong_df_subjects_en.pkl")
df.head()
```

```python
target = [column for column in df.columns if "subj_" in column]
X_train, X_test, y_train, y_test = train_test_split(df["cleaned_abstract"], df[target],
                                                   train_size=0.8, test_size=0.2,
                                                   random_state=42)

y_train_classes = {
    field: y_train[field] for field in target
}

clfs = {
    field: DummyClassifier(strategy="stratified") for field in target
}


from tqdm.notebook import tqdm
for field, clf in tqdm(clfs.items()):
    clf.fit(X_train, y_train_classes[field])
```

```python
from sklearn.metrics import (
    accuracy_score,
    recall_score,
    precision_score,
    f1_score
)

accuracy = {}
recall = {}
precision = {}
f1 = {}

for subj in tqdm(target):
    #print(subj.upper())
    y_pred = clfs[subj].predict(X_test)
    accuracy[subj] = accuracy_score(y_test[subj], y_pred)
    recall[subj] = recall_score(y_test[subj], y_pred, zero_division=0)
    precision[subj] = precision_score(y_test[subj], y_pred, zero_division=0)
    f1[subj] = f1_score(y_test[subj], y_pred, zero_division=0)
```

```python
print(f"Average accuracy: {np.mean(list(accuracy.values())):.4f}")
print(f"Average recall: {np.mean(list(recall.values())):.4f}")
print(f"Average precision: {np.mean(list(precision.values())):.4f}")
print(f"Average F1-score: {np.mean(list(f1.values())):.4f}")
```
